```python
import pandas as pd

df = pd.read_json("India_runs_data_and_ai_challenge/candidates.jsonl", lines=True)
df
```

```python
df.head()
```

```python
# Expand flat dict columns: profile + redrob_signals
df_profile = pd.json_normalize(df["profile"]).add_prefix("profile.")
df_signals = pd.json_normalize(df["redrob_signals"]).add_prefix("signals.")
df_flat = pd.concat([df[["candidate_id"]], df_profile, df_signals], axis=1)
df_flat.head()
```

```python
# Expand career_history list column — one row per job
df_careers = df[["candidate_id"]].join(
    df["career_history"].explode().apply(pd.Series).add_prefix("career.")
).reset_index(drop=True)
df_careers.head(10)
```

```python
# Expand skills list column — one row per skill
df_skills = df[["candidate_id"]].join(
    df["skills"].explode().apply(pd.Series).add_prefix("skill.")
).reset_index(drop=True)
df_skills.head(10)
```

```python
# Expand education list column — one row per degree
df_edu = df[["candidate_id"]].join(
    df["education"].explode().apply(pd.Series).add_prefix("edu.")
).reset_index(drop=True)
df_edu.head(10)
```

```python
# Expand languages list column — one row per language
df_languages = df[["languages"]].join(
    df["languages"].explode().apply(pd.Series).add_prefix("language.")
).reset_index(drop=True)
df_languages.head(10)
```

```python
# ---- 10. Honeypot Detection Heuristics ----
print("=== HONEYPOT DETECTION ANALYSIS ===\n")

# Heuristic 1: Expert skills with 0 or very low duration
expert_skills = df_skills[df_skills["skill.proficiency"] == "expert"]
expert_low_duration = expert_skills[expert_skills["skill.duration_months"] <= 3]
honeypot_h1 = expert_low_duration.groupby("candidate_id").size()
suspicious_h1 = honeypot_h1[honeypot_h1 >= 3]  # 3+ expert skills with <=3 months
print(f"H1: Candidates with 3+ 'expert' skills with ≤3 months usage: {len(suspicious_h1)}")
if len(suspicious_h1) > 0:
    print(f"    Top offenders: {suspicious_h1.sort_values(ascending=False).head(10).to_dict()}")

# Heuristic 2: Many expert skills overall (keyword stuffing)
expert_per_candidate = expert_skills.groupby("candidate_id").size()
many_experts = expert_per_candidate[expert_per_candidate >= 8]
print(f"\nH2: Candidates with 8+ expert-level skills: {len(many_experts)}")

# Heuristic 3: Title-career mismatch (non-tech title but AI skills)
non_tech_titles = df_flat[df_flat["profile.current_title"].str.contains(
    "Marketing|Sales|Accountant|HR Manager|Operations Manager|Customer Support|Teacher|Recruiter|Content Writer",
    case=False, na=False
)]
# Among these, how many have AI/ML skills?
non_tech_ids = set(non_tech_titles["candidate_id"])
ai_skill_names = ["Embeddings", "NLP", "Deep Learning", "Machine Learning", "PyTorch",
                   "TensorFlow", "LLM", "Fine-tuning", "RAG", "BERT", "Transformers"]
has_ai_skills = df_skills[
    (df_skills["candidate_id"].isin(non_tech_ids)) &
    (df_skills["skill.name"].str.contains("|".join(ai_skill_names), case=False, na=False))
]
mismatch_ids = has_ai_skills["candidate_id"].nunique()
print(f"\nH3: Non-tech title but has AI/ML skills (keyword stuffers): {mismatch_ids}")

# Heuristic 4: Experience years vs career history sum mismatch
career_total = df_careers.groupby("candidate_id")["career.duration_months"].sum() / 12.0
career_total.name = "career_total_years"
merged = df_flat[["candidate_id", "profile.years_of_experience"]].merge(
    career_total.reset_index(), on="candidate_id", how="left"
)
merged["gap"] = abs(merged["profile.years_of_experience"] - merged["career_total_years"])
big_gap = merged[merged["gap"] > 5]
print(f"\nH4: Candidates with >5yr gap between stated exp and career sum: {len(big_gap)}")

print(f"\n=== Total unique suspicious candidates (union of H1-H4): ===")
all_suspicious = set()
all_suspicious.update(suspicious_h1.index.tolist() if len(suspicious_h1) > 0 else [])
all_suspicious.update(many_experts.index.tolist())
all_suspicious.update(has_ai_skills["candidate_id"].unique().tolist())
all_suspicious.update(big_gap["candidate_id"].tolist())
print(f"  {len(all_suspicious)} candidates flagged")
```

```python
# ---- 12. "Ideal Candidate" Funnel Analysis ----
# How many candidates survive each filter layer?
print("=== CANDIDATE FUNNEL (progressive filtering) ===\n")

all_ids = set(df_flat["candidate_id"])
print(f"0. Total pool:                       {len(all_ids):>7,}")

# Filter 1: India-based
india_ids = set(df_flat.loc[df_flat["profile.country"] == "India", "candidate_id"])
print(f"1. Based in India:                   {len(india_ids):>7,}")

# Filter 2: 3-15 years experience
exp_mask = (df_flat["profile.years_of_experience"] >= 3) & (df_flat["profile.years_of_experience"] <= 15)
exp_ids = set(df_flat.loc[exp_mask, "candidate_id"])
f2 = india_ids & exp_ids
print(f"2. + 3-15 yrs experience:            {len(f2):>7,}")

# Filter 3: Has at least one AI/ML-relevant skill
ai_keywords = ["Python", "NLP", "Machine Learning", "Deep Learning", "PyTorch",
               "TensorFlow", "Embeddings", "BERT", "Transformers", "LLM",
               "Fine-tuning", "RAG", "Retrieval", "Ranking", "Search",
               "FAISS", "Elasticsearch", "Vector", "Recommendation",
               "scikit-learn", "XGBoost", "Neural", "Information Retrieval",
               "Pinecone", "Weaviate", "Qdrant", "Milvus", "OpenSearch",
               "Fine-tuning LLMs", "LoRA", "QLoRA", "BM25", "Hugging Face",
               "MLOps", "Sentence Transformers", "Data Science"]
ai_skill_mask = df_skills["skill.name"].str.contains("|".join(ai_keywords), case=False, na=False)
ai_skill_candidates = set(df_skills.loc[ai_skill_mask, "candidate_id"].unique())
f3 = f2 & ai_skill_candidates
print(f"3. + Has AI/ML-relevant skills:      {len(f3):>7,}")

# Filter 4: Tech-relevant current title
tech_title_mask = df_flat["profile.current_title"].str.contains(
    "Engineer|Developer|Scientist|Architect|Analyst|ML|AI|Data|Research|Platform|DevOps|SRE|CTO|VP Eng",
    case=False, na=False
)
tech_ids = set(df_flat.loc[tech_title_mask, "candidate_id"])
f4 = f3 & tech_ids
print(f"4. + Tech-relevant current title:    {len(f4):>7,}")

# Filter 5: Has some product company experience (not consulting-only)
consulting_set = {"TCS", "Infosys", "Wipro", "Accenture", "Cognizant", "Capgemini", "HCL", "Tech Mahindra"}
career_companies = df_careers.groupby("candidate_id")["career.company"].apply(set)
consulting_only = career_companies[career_companies.apply(lambda cos: cos.issubset(consulting_set | {"Mindtree", "Mphasis", "LTI", "L&T Infotech"}))]
consulting_only_ids = set(consulting_only.index)
f5 = f4 - consulting_only_ids
print(f"5. + Not consulting-only career:     {len(f5):>7,}")

# Filter 6: Active recently (last 90 days) or open to work
recent_or_otw = set(df_flat.loc[
    (df_flat["days_since_active"] <= 90) | (df_flat["signals.open_to_work_flag"] == True),
    "candidate_id"
])
f6 = f5 & recent_or_otw
print(f"6. + Active <90d or open to work:    {len(f6):>7,}")

# Filter 7: Multiple AI/ML skills (not just Python)
multi_ai = df_skills[
    (df_skills["candidate_id"].isin(f6)) &
    (df_skills["skill.name"].str.contains(
        "NLP|Machine Learning|Deep Learning|PyTorch|TensorFlow|Embeddings|BERT|Transformers|LLM|"
        "Fine-tuning|RAG|Retrieval|Ranking|FAISS|Elasticsearch|Vector|Recommendation|"
        "scikit-learn|XGBoost|Neural|Pinecone|Weaviate|Qdrant|MLOps|Hugging Face|BM25|LoRA",
        case=False, na=False
    ))
].groupby("candidate_id").size()
multi_ai_ids = set(multi_ai[multi_ai >= 3].index)
f7 = f6 & multi_ai_ids
print(f"7. + 3+ AI/ML skills (not just Py):  {len(f7):>7,}")

print(f"\n=== Final shortlist for deep analysis: {len(f7):,} candidates ===")
```

```python
# ---- 13. Deep-dive: Look at a few strong candidates from the funnel ----
# Show full profiles of top candidates from f7 to understand what "good" looks like

shortlist = df_flat[df_flat["candidate_id"].isin(f7)].copy()

# Sort by a rough relevance proxy: years_of_exp in range + github + response rate
shortlist["rough_score"] = (
    shortlist["signals.recruiter_response_rate"].fillna(0) * 20 +
    shortlist["signals.github_activity_score"].clip(lower=0).fillna(0) * 0.3 +
    shortlist["signals.profile_completeness_score"].fillna(0) * 0.1 +
    (shortlist["signals.notice_period_days"] <= 30).astype(int) * 10 -
    shortlist["days_since_active"].fillna(999) * 0.05
)

top_candidates = shortlist.nlargest(10, "rough_score")

for _, row in top_candidates.iterrows():
    cid = row["candidate_id"]
    print(f"\n{'='*80}")
    print(f"  {cid} | {row['profile.current_title']} at {row['profile.current_company']}")
    print(f"  {row['profile.headline']}")
    print(f"  Location: {row['profile.location']}, {row['profile.country']} | YoE: {row['profile.years_of_experience']}")
    print(f"  Industry: {row['profile.current_industry']} | Company Size: {row['profile.current_company_size']}")
    
    # Skills
    cand_skills = df_skills[df_skills["candidate_id"] == cid][["skill.name", "skill.proficiency", "skill.duration_months"]]
    print(f"  Skills: {', '.join(cand_skills['skill.name'].tolist())}")
    
    # Key signals
    print(f"  Response Rate: {row['signals.recruiter_response_rate']:.2f} | "
          f"GitHub: {row['signals.github_activity_score']} | "
          f"Notice: {row['signals.notice_period_days']}d | "
          f"Open to Work: {row['signals.open_to_work_flag']} | "
          f"Days Inactive: {row['days_since_active']}")
    
    # Career summary
    cand_careers = df_careers[df_careers["candidate_id"] == cid][["career.company", "career.title", "career.duration_months", "career.industry"]]
    for _, c in cand_careers.iterrows():
        print(f"    → {c['career.title']} at {c['career.company']} ({c['career.duration_months']}mo, {c['career.industry']})")
    
    print(f"  Rough Score: {row['rough_score']:.1f}")
```

# ===== SKILLS vs CAREER DESCRIPTION MISMATCH ANALYSIS =====
# Finding candidates whose skills section has NO ML/AI keywords,
# but whose career descriptions reveal relevant ML/AI/retrieval experience.
# These are the "plain-language Tier 5" candidates the JD warns about.

```python
import re
from collections import Counter

# ---- Step 1: Define ML/AI keyword sets ----

# Skills-section keywords (exact skill names from the dataset)
SKILL_ML_KEYWORDS = [
    "NLP", "Machine Learning", "Deep Learning", "PyTorch", "TensorFlow",
    "Embeddings", "BERT", "Transformers", "LLM", "LLMs",
    "Fine-tuning", "Fine-tuning LLMs", "RAG", "Retrieval",
    "FAISS", "Elasticsearch", "Pinecone", "Weaviate", "Qdrant", "Milvus",
    "OpenSearch", "scikit-learn", "XGBoost", "Neural Networks",
    "Recommendation Systems", "Recommendation System",
    "Information Retrieval", "Learning to Rank", "BM25",
    "Sentence Transformers", "Hugging Face", "MLOps",
    "LoRA", "QLoRA", "PEFT", "LangChain", "LlamaIndex", "pgvector",
    "Computer Vision", "Image Classification", "Object Detection",
    "Speech Recognition", "ASR", "TTS",
    "Reinforcement Learning", "Data Science", "Feature Engineering",
    "Semantic Search", "Vector Search", "Prompt Engineering",
    "Diffusion Models", "CNN", "RNN", "GAN",
    "Statistical Modeling", "Haystack", "BentoML",
    "Weights & Biases", "MLflow", "Kubeflow",
    "YOLO", "OpenCV",
]
skill_ml_pattern = "|".join([re.escape(s) for s in SKILL_ML_KEYWORDS])

# Career description keywords — broader, includes plain-language phrases
# that indicate ML/AI work even without using exact tool names
CAREER_ML_PATTERNS = {
    # Core ML/AI work
    "model training/building": r"\b(trained|training|built|build|fine[- ]?tun|finetuned)\b.{0,30}\b(model|classifier|regressor|neural|network|transformer|llm|bert)",
    "ml model references": r"\b(machine learning|deep learning|ml model|neural net|prediction model|classification model|regression model|language model)",
    "embeddings/vectors": r"\b(embedding|vector|semantic|dense retrieval|similarity search|cosine sim|nearest neighbor|knn|ann\b|faiss|approximate nearest)",
    "retrieval/ranking/search": r"\b(retrieval|ranking|re[- ]?rank|search relevance|search quality|information retrieval|learning.to.rank|bm25|tf[- ]?idf|inverted index|candidate retrieval)",
    "recommendation systems": r"\b(recommend|collaborative filter|content[- ]?based filter|matrix factorization|user[- ]?item|click[- ]?through|personali[sz])",
    "nlp work": r"\b(natural language|nlp|text classification|named entity|sentiment|tokeniz|word2vec|seq2seq|attention mechanism|transformer)",
    "production ml": r"\b(deployed.{0,20}(model|pipeline|system)|model serving|model inference|production.{0,15}(ml|model|pipeline)|a/b test.{0,15}(model|rank)|online.{0,10}(serving|inference|evaluation))",
    "eval frameworks": r"\b(ndcg|mrr|mean average precision|precision.at|recall.at|f1.score|auc[- ]?roc|evaluation.{0,15}(framework|pipeline|metric)|offline.{0,10}eval)",
    "vector db work": r"\b(pinecone|weaviate|qdrant|milvus|chroma|elasticsearch|opensearch|vector.{0,5}(database|db|store|index)|faiss|annoy)",
    "llm/genai work": r"\b(llm|large language|gpt|claude|gemini|prompt engineer|chain.of.thought|rag|retrieval.augmented|fine[- ]?tun.{0,10}(llm|gpt|model|language))",
    "data pipeline + ml": r"\b(feature.{0,5}(engineering|pipeline|store|extraction)|training.{0,5}(pipeline|data|loop)|data.{0,5}(pipeline|preprocessing).{0,30}(model|ml|predict))",
    "experiment tracking": r"\b(mlflow|wandb|weights.{0,3}biases|experiment.{0,5}track|hyperparameter.{0,5}(tun|search|optim)|model.{0,5}registry)",
}

print(f"Skills ML keywords: {len(SKILL_ML_KEYWORDS)}")
print(f"Career description pattern categories: {len(CAREER_ML_PATTERNS)}")
print("Ready for mismatch analysis.")
```

```python
# ---- Step 2: Identify candidates with NO ML/AI skills in skills section ----

# For each candidate, check if ANY of their skills match ML keywords
candidates_with_ml_skills = df_skills.loc[
    df_skills["skill.name"].str.contains(skill_ml_pattern, case=False, na=False),
    "candidate_id"
].unique()

all_candidate_ids = set(df["candidate_id"])
candidates_without_ml_skills = all_candidate_ids - set(candidates_with_ml_skills)

print(f"Candidates WITH at least 1 ML/AI skill in skills section:    {len(candidates_with_ml_skills):>7,}")
print(f"Candidates with ZERO ML/AI skills in skills section:         {len(candidates_without_ml_skills):>7,}")

# What are the top skills for these "no ML skills" candidates?
no_ml_skills_data = df_skills[df_skills["candidate_id"].isin(candidates_without_ml_skills)]
print(f"\nTop 20 skills among candidates with NO ML skills listed:")
print(no_ml_skills_data["skill.name"].value_counts().head(20).to_string())
```

```python
# ---- Step 3: Scan career descriptions for ML/AI work ----
# Concatenate all career descriptions per candidate

career_text_per_candidate = (
    df_careers
    .groupby("candidate_id")["career.description"]
    .apply(lambda descs: " ".join(descs.dropna().tolist()).lower())
    .reset_index()
    .rename(columns={"career.description": "all_career_text"})
)

# Also grab career titles concatenated
career_titles_per_candidate = (
    df_careers
    .groupby("candidate_id")["career.title"]
    .apply(lambda titles: " | ".join(titles.dropna().tolist()).lower())
    .reset_index()
    .rename(columns={"career.title": "all_career_titles"})
)

career_text_per_candidate = career_text_per_candidate.merge(
    career_titles_per_candidate, on="candidate_id"
)

# Also include profile summary + headline
profile_text = df_flat[["candidate_id", "profile.summary", "profile.headline"]].copy()
profile_text["profile_text"] = (
    profile_text["profile.summary"].fillna("") + " " + profile_text["profile.headline"].fillna("")
).str.lower()

career_text_per_candidate = career_text_per_candidate.merge(
    profile_text[["candidate_id", "profile_text"]], on="candidate_id"
)

# Combine all text
career_text_per_candidate["full_text"] = (
    career_text_per_candidate["all_career_text"] + " " +
    career_text_per_candidate["all_career_titles"] + " " +
    career_text_per_candidate["profile_text"]
)

print(f"Built text corpus for {len(career_text_per_candidate):,} candidates")
print(f"Average text length: {career_text_per_candidate['full_text'].str.len().mean():.0f} chars")
```

```python
# ---- Step 4: Find the mismatch — NO ML skills listed, but career text shows ML work ----

# Filter to candidates without ML skills
no_ml_text = career_text_per_candidate[
    career_text_per_candidate["candidate_id"].isin(candidates_without_ml_skills)
].copy()

print(f"Scanning {len(no_ml_text):,} candidates (no ML skills listed) for ML keywords in career text...\n")

# Score each candidate's text against every pattern category
pattern_hits = {}
for category, pattern in CAREER_ML_PATTERNS.items():
    matches = no_ml_text["full_text"].str.contains(pattern, case=False, na=False, regex=True)
    no_ml_text[f"hit_{category}"] = matches
    hit_count = matches.sum()
    pattern_hits[category] = hit_count
    print(f"  {category:40s}: {hit_count:>6,} candidates match")

# Count total pattern categories matched per candidate
hit_cols = [c for c in no_ml_text.columns if c.startswith("hit_")]
no_ml_text["total_ml_categories_hit"] = no_ml_text[hit_cols].sum(axis=1)

# Candidates with at least 1 ML pattern in career text
has_ml_in_career = no_ml_text[no_ml_text["total_ml_categories_hit"] >= 1]
has_strong_ml = no_ml_text[no_ml_text["total_ml_categories_hit"] >= 3]

print(f"\n=== MISMATCH SUMMARY ===")
print(f"No ML skills listed, but ≥1 ML pattern in career text:  {len(has_ml_in_career):,}")
print(f"No ML skills listed, but ≥3 ML patterns in career text: {len(has_strong_ml):,} (STRONG mismatch)")
print(f"No ML skills listed AND no ML in career text:           {len(no_ml_text) - len(has_ml_in_career):,}")
```

```python
# ---- Step 6: Deep-dive into the strongest mismatch candidates ----
# These are "hidden gems" — no ML skills listed, but career text shows real ML work

# Merge with profile data for richer view
strong_mismatch = has_strong_ml.merge(
    df_flat[["candidate_id", "profile.current_title", "profile.current_company",
             "profile.headline", "profile.location", "profile.country",
             "profile.years_of_experience", "profile.current_industry",
             "signals.recruiter_response_rate", "signals.github_activity_score",
             "signals.notice_period_days", "signals.open_to_work_flag"]],
    on="candidate_id"
)

# Sort by number of ML categories matched
strong_mismatch = strong_mismatch.sort_values("total_ml_categories_hit", ascending=False)

print(f"=== TOP 15 'HIDDEN GEM' CANDIDATES ===")
print(f"(No ML/AI skills in skills section, but ≥3 ML patterns in career text)\n")

for idx, (_, row) in enumerate(strong_mismatch.head(15).iterrows()):
    cid = row["candidate_id"]
    print(f"\n{'='*90}")
    print(f"  #{idx+1} | {cid} | ML categories matched: {row['total_ml_categories_hit']}")
    print(f"  Title: {row['profile.current_title']} at {row['profile.current_company']}")
    print(f"  Headline: {row['profile.headline']}")
    print(f"  Location: {row['profile.location']}, {row['profile.country']} | YoE: {row['profile.years_of_experience']}")
    print(f"  Industry: {row['profile.current_industry']}")
    print(f"  Signals: Response Rate={row['signals.recruiter_response_rate']:.2f}, "
          f"GitHub={row['signals.github_activity_score']}, "
          f"Notice={row['signals.notice_period_days']}d, "
          f"Open to Work={row['signals.open_to_work_flag']}")

    # Which ML categories matched
    matched_cats = [c.replace("hit_", "") for c in hit_cols if row[c]]
    print(f"  ML patterns found: {', '.join(matched_cats)}")

    # Their actual skills
    cand_skills = df_skills[df_skills["candidate_id"] == cid]["skill.name"].tolist()
    print(f"  Listed skills: {', '.join(cand_skills)}")

    # Career history
    cand_careers = df_careers[df_careers["candidate_id"] == cid]
    for _, c in cand_careers.iterrows():
        print(f"    -> {c['career.title']} at {c['career.company']} ({c['career.duration_months']}mo)")
        desc = c['career.description']
        if pd.notna(desc):
            # Highlight ML-relevant phrases in the description
            print(f"       {desc[:200]}{'...' if len(str(desc)) > 200 else ''}")
```

```python
# ---- Step 7: Do the REVERSE check too — ML skills listed but NO ML in career text ----
# These are potential keyword stuffers / honeypots

has_ml_text = career_text_per_candidate[
    career_text_per_candidate["candidate_id"].isin(candidates_with_ml_skills)
].copy()

# Score career text
for category, pattern in CAREER_ML_PATTERNS.items():
    has_ml_text[f"hit_{category}"] = has_ml_text["full_text"].str.contains(
        pattern, case=False, na=False, regex=True
    )

hit_cols_rev = [c for c in has_ml_text.columns if c.startswith("hit_")]
has_ml_text["total_ml_categories_hit"] = has_ml_text[hit_cols_rev].sum(axis=1)

# Keyword stuffers: have ML skills but ZERO ML patterns in career text
keyword_stuffers = has_ml_text[has_ml_text["total_ml_categories_hit"] == 0]

print(f"=== REVERSE MISMATCH (Keyword Stuffers) ===")
print(f"Have ML skills in skills section:                    {len(has_ml_text):,}")
print(f"  but ZERO ML patterns in career text (stuffers):   {len(keyword_stuffers):,}")
print(f"  with ≥1 ML pattern in career text (legitimate):   {len(has_ml_text) - len(keyword_stuffers):,}")

# How many ML skills do keyword stuffers have?
stuffer_ids = set(keyword_stuffers["candidate_id"])
stuffer_skills = df_skills[
    (df_skills["candidate_id"].isin(stuffer_ids)) &
    (df_skills["skill.name"].str.contains(skill_ml_pattern, case=False, na=False))
]
ml_skills_per_stuffer = stuffer_skills.groupby("candidate_id").size()
print(f"\nML skills per keyword stuffer:")
print(ml_skills_per_stuffer.describe().to_string())

# Show a few keyword stuffers
print(f"\n--- Sample Keyword Stuffers (ML skills listed, but career text has no ML work) ---")
stuffer_sample = keyword_stuffers.head(5)
for _, row in stuffer_sample.iterrows():
    cid = row["candidate_id"]
    profile = df_flat[df_flat["candidate_id"] == cid].iloc[0]
    cand_skills = df_skills[
        (df_skills["candidate_id"] == cid) &
        (df_skills["skill.name"].str.contains(skill_ml_pattern, case=False, na=False))
    ]["skill.name"].tolist()
    print(f"\n  {cid} | {profile['profile.current_title']} at {profile['profile.current_company']}")
    print(f"  ML skills listed: {', '.join(cand_skills)}")
    print(f"  Career text: {row['all_career_text'][:250]}...")
```
