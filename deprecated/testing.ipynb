```python
import pandas as pd

df = pd.read_json("candidates.jsonl", lines=True)
df
```

```python
df.head()
```

```python
# Expand flat dict columns: profile + redrob_signals
df_profile = pd.json_normalize(df["profile"]).add_prefix("profile.")
df_signals = pd.json_normalize(df["redrob_signals"]).add_prefix("signals.")
df_flat = pd.concat([df[["candidate_id"]], df_profile, df_signals], axis=1)
df_flat.head()
```

```python
# Expand career_history list column — one row per job
df_careers = df[["candidate_id"]].join(
    df["career_history"].explode().apply(pd.Series).add_prefix("career.")
).reset_index(drop=True)
df_careers.head(10)
```

```python
# Expand skills list column — one row per skill
df_skills = df[["candidate_id"]].join(
    df["skills"].explode().apply(pd.Series).add_prefix("skill.")
).reset_index(drop=True)
df_skills.head(10)
```

```python
# Expand education list column — one row per degree
df_edu = df[["candidate_id"]].join(
    df["education"].explode().apply(pd.Series).add_prefix("edu.")
).reset_index(drop=True)
df_edu.head(10)
```

```python
# Expand languages list column — one row per language
df_languages = df[["languages"]].join(
    df["languages"].explode().apply(pd.Series).add_prefix("language.")
).reset_index(drop=True)
df_languages.head(10)
```
