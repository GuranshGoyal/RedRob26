```python
import pandas as pd
import ast

# Load the data
df = pd.read_csv('dataset.csv')

# Check the datatype of skills column
print("Skills column dtype:", df['skills'].dtype)
print("\nSample skills value:")
print(df['skills'].iloc[0])

# Parse the skills column (convert string to list)
df['skills_parsed'] = df['skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Extract skill names from parsed skills
df['skill_names'] = df['skills_parsed'].apply(lambda x: [skill['name'] for skill in x])

# Define relevant skills
relevant_skills = {
    "Databricks", "Terraform", "Kafka", "BigQuery", "CI/CD", "Airflow",
    "Flask", "AWS", "Kubernetes", "ETL", "Docker", "Apache Beam",
    "Go", "Java", "TypeScript", "JavaScript", "dbt", "REST APIs",
    "Spark", "Snowflake", "GCP", "PostgreSQL", "Rust", "Apache Flink",
    "gRPC", "SQL", "Hadoop", "Redis", "FastAPI", "Microservices",
    "Data Pipelines", "Azure",
    "Hugging Face Transformers", "LangChain", "Information Retrieval",
    "LLMs", "Recommendation Systems", "Semantic Search",
    "Sentence Transformers", "Embeddings", "Vector Search",
    "Pinecone", "Prompt Engineering", "FAISS", "RAG",
    "Fine-tuning LLMs", "YOLO", "GANs", "Feature Engineering",
    "OpenCV", "ASR", "Image Classification", "Computer Vision",
    "Speech Recognition", "CNN", "Kubeflow", "MLOps",
    "BentoML", "Reinforcement Learning", "Data Science",
    "Object Detection", "Diffusion Models", "MLflow",
    "Time Series", "Weights & Biases", "Forecasting",
    "TTS", "Statistical Modeling", "QLoRA", "pgvector",
    "Weaviate", "Milvus", "Learning to Rank", "BM25",
    "TensorFlow", "Qdrant", "Python", "PyTorch",
    "PEFT", "LoRA", "NLP", "Machine Learning",
    "Deep Learning", "Haystack", "Elasticsearch",
    "LlamaIndex", "scikit-learn", "OpenSearch",
    "Information Retrieval Systems", "Text Encoders",
    "Search Backend", "Model Adaptation",
    "Search & Discovery", "Ranking Systems",
    "Vector Representations", "Content Matching",
    "Workflow Orchestration", "Search Infrastructure",
    "Indexing Algorithms", "Open-source ML libraries",
    "Natural Language Processing", "Document Processing"
}

# Filter rows that have at least 1 relevant skill with duration > 0
df_filtered = df[df['skills_parsed'].apply(
    lambda skills: any(
        skill['name'] in relevant_skills and skill.get('duration_months', 0) > 0 
        for skill in skills
    )
)]

print(f"\nOriginal rows: {len(df)}")
print(f"Filtered rows: {len(df_filtered)}")
print(f"Rows removed: {len(df) - len(df_filtered)}")

# Display sample of filtered data
print("\nSample filtered rows:")
print(df_filtered[['candidate_id', 'skill_names']].head())
```

```python
# Parse profile and redrob_signals columns
df['profile_parsed'] = df['profile'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})
df['redrob_signals_parsed'] = df['redrob_signals'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})

# Extract location and willing_to_relocate
df['location'] = df['profile_parsed'].apply(lambda x: x.get('location', ''))
df['willing_to_relocate'] = df['redrob_signals_parsed'].apply(lambda x: x.get('willing_to_relocate', False))

# Filter with all three conditions:
# 1. At least 1 relevant skill with duration > 0
# 2. Location is Noida or Pune OR willing_to_relocate is True
df_filtered = df[df.apply(
    lambda row: (
        any(skill['name'] in relevant_skills and skill.get('duration_months', 0) > 0 
            for skill in row['skills_parsed'])
        and (row['location'] in ['Noida', 'Pune'] or row['willing_to_relocate'])
    ), axis=1
)]

print(f"Filtered rows: {len(df_filtered)}")
```

```python
# Parse career_history column
df['career_history_parsed'] = df['career_history'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Function to check if candidate should be eliminated based on point 3
def should_eliminate(career_history):
    if len(career_history) <= 1:
        return True 
    # Check if any job has duration < 18 months
    return any(job.get('duration_months', 0) < 18 for job in career_history)

# Filter with all conditions:
# 1. At least 1 relevant skill with duration > 0
# 2. Location is Noida/Pune OR willing_to_relocate is True
# 3. NOT (if more than 1 job, then any job duration < 18 months)
df_filtered = df[df.apply(
    lambda row: (
        any(skill['name'] in relevant_skills and skill.get('duration_months', 0) > 0 
            for skill in row['skills_parsed'])
        and (row['location'] in ['Noida', 'Pune'] or row['willing_to_relocate'])
        and not should_eliminate(row['career_history_parsed'])
    ), axis=1
)]
print(f"Filtered rows: {len(df_filtered)}")
```

```python
import pandas as pd
import ast
from collections import Counter
# import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dataset.csv')

# Parse profile and career_history columns
df['profile_parsed'] = df['profile'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})
df['career_history_parsed'] = df['career_history'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Extract current companies from profile
current_companies = df['profile_parsed'].apply(lambda x: x.get('current_company', ''))
current_industries = df['profile_parsed'].apply(lambda x: x.get('current_industry', ''))
current_company_sizes = df['profile_parsed'].apply(lambda x: x.get('current_company_size', ''))

# Extract all companies from career_history
all_companies = []
all_industries = []
all_company_sizes = []

for career in df['career_history_parsed']:
    for job in career:
        all_companies.append(job.get('company', ''))
        all_industries.append(job.get('industry', ''))
        all_company_sizes.append(job.get('company_size', ''))

print("=" * 80)
print("COMPANY AND DOMAIN ANALYSIS")
print("=" * 80)

# 1. Current Companies Analysis
print("\n1. CURRENT COMPANIES (from profile)")
print("-" * 80)
current_companies_clean = [c for c in current_companies if c and c != '']
print(f"Total unique current companies: {len(set(current_companies_clean))}")
print(f"Total candidates with current company info: {len(current_companies_clean)}")

print("\nTop 20 Current Companies by Frequency:")
current_company_counts = Counter(current_companies_clean)
for company, count in current_company_counts.most_common(20):
    print(f"  {company}: {count} candidates")

# 2. Career History Companies Analysis
print("\n2. ALL COMPANIES (from career_history)")
print("-" * 80)
all_companies_clean = [c for c in all_companies if c and c != '']
print(f"Total unique companies in career history: {len(set(all_companies_clean))}")
print(f"Total company entries: {len(all_companies_clean)}")

print("\nTop 20 Companies by Frequency (Career History):")
company_counts = Counter(all_companies_clean)
for company, count in company_counts.most_common(20):
    print(f"  {company}: {count} entries")

# 3. Industry/Domain Analysis
print("\n3. INDUSTRY/DOMAIN ANALYSIS")
print("-" * 80)

current_industries_clean = [i for i in current_industries if i and i != '']
print(f"\nCurrent Industries - Unique count: {len(set(current_industries_clean))}")
print("\nTop 15 Current Industries:")
current_industry_counts = Counter(current_industries_clean)
for industry, count in current_industry_counts.most_common(15):
    print(f"  {industry}: {count} candidates")

all_industries_clean = [i for i in all_industries if i and i != '']
print(f"\nCareer History Industries - Unique count: {len(set(all_industries_clean))}")
print("\nTop 15 Industries (Career History):")
industry_counts = Counter(all_industries_clean)
for industry, count in industry_counts.most_common(15):
    print(f"  {industry}: {count} entries")

# 4. Company Size Analysis
print("\n4. COMPANY SIZE ANALYSIS")
print("-" * 80)

current_sizes_clean = [s for s in current_company_sizes if s and s != '']
print(f"\nCurrent Company Sizes Distribution:")
current_size_counts = Counter(current_sizes_clean)
for size, count in sorted(current_size_counts.items()):
    print(f"  {size}: {count} candidates")

all_sizes_clean = [s for s in all_company_sizes if s and s != '']
print(f"\nCareer History Company Sizes Distribution:")
size_counts = Counter(all_sizes_clean)
for size, count in sorted(size_counts.items()):
    print(f"  {size}: {count} entries")

# 5. Company-Industry Mapping
print("\n5. COMPANY-INDUSTRY MAPPING (Top Companies)")
print("-" * 80)

# Create mapping of company to industry
company_industry_map = {}
for career in df['career_history_parsed']:
    for job in career:
        company = job.get('company', '')
        industry = job.get('industry', '')
        if company and industry:
            if company not in company_industry_map:
                company_industry_map[company] = Counter()
            company_industry_map[company][industry] += 1

print("\nTop 15 Companies with their Industry Distribution:")
for company in list(company_counts.most_common(15)):
    company_name = company[0]
    if company_name in company_industry_map:
        print(f"\n  {company_name}:")
        for industry, count in company_industry_map[company_name].most_common():
            print(f"    - {industry}: {count}")

# 6. Industry Diversity Analysis
print("\n6. INDUSTRY DIVERSITY BY COMPANY")
print("-" * 80)

company_diversity = {}
for company, industries in company_industry_map.items():
    company_diversity[company] = len(industries)

print("\nTop 15 Companies by Industry Diversity:")
for company, diversity in sorted(company_diversity.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(f"  {company}: {diversity} different industries")

# 7. Candidate Company Count Analysis
print("\n7. CANDIDATE CAREER HISTORY DEPTH")
print("-" * 80)

job_counts = [len(career) for career in df['career_history_parsed']]
print(f"Average jobs per candidate: {sum(job_counts)/len(job_counts):.2f}")
print(f"Min jobs: {min(job_counts)}")
print(f"Max jobs: {max(job_counts)}")

print("\nJob Count Distribution:")
job_count_dist = Counter(job_counts)
for count, freq in sorted(job_count_dist.items()):
    print(f"  {count} jobs: {freq} candidates")

# 8. Company Domain Categories
print("\n8. COMPANY DOMAIN CATEGORIZATION")
print("-" * 80)

# Categorize companies by domain patterns
tech_companies = []
it_services = []
conglomerates = []
manufacturing = []
finance = []
consulting = []
other = []

for company in set(all_companies_clean):
    company_lower = company.lower()
    if any(keyword in company_lower for keyword in ['tech', 'software', 'data', 'cloud', 'ai', 'ml', 'pied piper', 'hooli']):
        tech_companies.append(company)
    elif any(keyword in company_lower for keyword in ['infosys', 'tcs', 'wipro', 'mindtree', 'accenture', 'cognizant']):
        it_services.append(company)
    elif any(keyword in company_lower for keyword in ['enterprises', 'conglomerate', 'stark', 'wayne']):
        conglomerates.append(company)
    elif any(keyword in company_lower for keyword in ['manufacturing', 'industrial', 'steel', 'auto']):
        manufacturing.append(company)
    elif any(keyword in company_lower for keyword in ['bank', 'financial', 'finance', 'investment']):
        finance.append(company)
    elif any(keyword in company_lower for keyword in ['consulting', 'mckinsey', 'bain', 'bcg']):
        consulting.append(company)
    else:
        other.append(company)

print(f"\nTech Companies: {len(tech_companies)}")
print(f"IT Services: {len(it_services)}")
print(f"Conglomerates: {len(conglomerates)}")
print(f"Manufacturing: {len(manufacturing)}")
print(f"Finance: {len(finance)}")
print(f"Consulting: {len(consulting)}")
print(f"Other/Diverse: {len(other)}")

print("\nSample companies by category:")
print(f"\nTech: {', '.join(tech_companies[:10])}")
print(f"\nIT Services: {', '.join(it_services[:10])}")
print(f"\nConglomerates: {', '.join(conglomerates[:10])}")

print("\n" + "=" * 80)
print("ANALYSIS COMPLETE")
print("=" * 80)
```

```python
import pandas as pd
import ast
from collections import Counter

# Load the dataset
df = pd.read_csv('dataset.csv')

# Parse profile and career_history columns
df['profile_parsed'] = df['profile'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})
df['career_history_parsed'] = df['career_history'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})

# Extract companies from profile (current companies with IT Services industry)
current_it_companies = []
for profile in df['profile_parsed']:
    company = profile.get('current_company', '')
    industry = profile.get('current_industry', '')
    if industry == 'IT Services' and company:
        current_it_companies.append(company)

# Extract all companies from career_history with IT Services industry
career_it_companies = []
for career in df['career_history_parsed']:
    for job in career:
        company = job.get('company', '')
        industry = job.get('industry', '')
        if industry == 'IT Services' and company:
            career_it_companies.append(company)

print("=" * 80)
print("IT SERVICES COMPANIES")
print("=" * 80)

# Current IT companies
print("\n1. CURRENT IT SERVICES COMPANIES (from profile)")
print("-" * 80)
current_it_counts = Counter(current_it_companies)
print(f"Total unique current IT companies: {len(current_it_counts)}")
print(f"Total entries: {len(current_it_companies)}")

print("\nCurrent IT Companies by Frequency:")
for company, count in current_it_counts.most_common():
    print(f"  {company}: {count} candidates")

# Career history IT companies
print("\n2. ALL IT SERVICES COMPANIES (from career_history)")
print("-" * 80)
career_it_counts = Counter(career_it_companies)
print(f"Total unique IT companies in career history: {len(career_it_counts)}")
print(f"Total entries: {len(career_it_companies)}")

print("\nIT Companies by Frequency (Career History):")
for company, count in career_it_counts.most_common():
    print(f"  {company}: {count} entries")

# Combined unique list
print("\n3. COMPLETE LIST OF UNIQUE IT SERVICES COMPANIES")
print("-" * 80)
all_it_companies = set(current_it_companies + career_it_companies)
print(f"Total unique IT companies: {len(all_it_companies)}")
print("\nAlphabetical list:")
for company in sorted(all_it_companies):
    print(f"  - {company}")

print("\n" + "=" * 80)
```

```python
# Define consulting firms list
consulting_firms = {
    'TCS', 'Infosys', 'Wipro', 'Accenture', 'Cognizant', 'Capgemini',
    'Mindtree', 'HCL Technologies', 'Tech Mahindra', 'Mphasis',
    'L&T Infotech', 'Hexaware', 'Zensar', 'KPIT', 'Birlasoft'
}

# Function to check if candidate has only worked at consulting firms
def only_consulting_firms(career_history):
    if not career_history:
        return False  # No career history, don't eliminate
    # Check if all companies in career are consulting firms
    all_companies = [job.get('company', '') for job in career_history]
    return all(company in consulting_firms for company in all_companies if company)

# Update the filter with all conditions:
# 1. At least 1 relevant skill with duration > 0
# 2. Location is Noida/Pune OR willing_to_relocate is True
# 3. NOT (more than 1 job AND any job duration < 18 months)
# 4. NOT (only worked at consulting firms)
df_filtered = df_filtered[df_filtered.apply(
    lambda row: (
        not only_consulting_firms(row['career_history_parsed'])
    ), axis=1
)]


print(f"Filtered rows: {len(df_filtered)}")
```
